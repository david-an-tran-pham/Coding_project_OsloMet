```python
# task 6: simple linear
# Already used statsmodel in task 4.

# matrix comparison 

y_asking=house_price[['askingPrice(million)']]


# from dataframe to array
array_sqm=np.array(x1)
array_asking=np.array(y_asking)




# x = (1,x1), (1,x2), ...(1,xn) 
x_sqm=np.matrix([np.ones(len(array_sqm)),array_sqm[:,0]]).T

# y = y1, y2, ..., yn
y_sqm=np.matrix(array_asking[:,0]).T

# A = (X^T*X)^-1*X^T*Y
A_sqm=np.linalg.inv(x_sqm.T.dot(x_sqm)).dot(x_sqm.T).dot(y_sqm)

# X^TX  = 2x100(matrix x)*100*2 (matrix y) = 2x2 matrix

# b is the constant or intercept
b_sqm=A_sqm[0,0]

# m is the coefficent
m_sqm=A_sqm[1,0]

# start, stop, number of samples
xx_sqm=np.linspace(0.0,300,num=100)

# Y = a + b*x
yy_sqm=np.array(b_sqm + m_sqm*xx_sqm)

# plotting:
plt.plot(xx_sqm,yy_sqm, color='g')
plt.scatter(array_sqm[:,0],array_asking[:,0],color='b')
plt.title('Matrix linear regression')
plt.ylabel('Asking price millions')
plt.xlabel('Square meter')
plt.show()
print('b=',format(b_sqm,'.3f'),'m=',format(m_sqm,'.3f'))




# finding sum of square error matrix
asking_array=np.array(house_price.loc[:,'askingPrice(million)'])

# using linear equation of square meter: y= -1.955 + 0.1075*x


y_asking_predict=np.array(b_sqm + m_sqm*x1)
y_asking_actual=asking_array
sse_list=[]

# Calculating Sum Squared Error:
for k in range(len(y_asking_predict)):
    sse =(y_asking_actual[k]-y_asking_predict[k]) 
    sse_list.append(sse)
        

# from list to array
sse_list_a=np.array(sse_list)

# calculating SSE
error_matrix=np.sum(sse_list_a.T.dot(sse_list_a))


# calculating mean squared error:
mean_square=error_matrix/len(y_asking_predict)

print('Sum Squared Error:',format(error_matrix,'.3f'))
print('Mean squared error:', format(mean_square,'.3f'))
```
